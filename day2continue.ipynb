```python
l = [1,2,3,4,4,"sudh","kmumar",324,34.456,"abc"]
```

```python
l1_num = []
l2_str = []
for i in l:
  if type(i) == int or type(i) == float:
    l1_num.append(i)
  else:
    l2_str.append(i)
```

```python
l3 = [1,2,3,6545.54,"adil","raza",True]
```

```python
l11_num = []
l22_str = []
l33_bool = []
for i in l3:
  if type(i) == int or type(i) == float:
    l11_num.append(i)
  elif type(i) == str:
    l22_str.append(i)
  else:
    l33_bool.append(i)
```

```python
num = int(input("Enter the number:"))
sum = 0
temp = num
while(temp != 0):
  ld = temp % 10
  sum = (sum * 10) + ld
  temp = temp // 10
if (sum == num):
  print("Palindrome")
else:
  print("not palindrome")
```

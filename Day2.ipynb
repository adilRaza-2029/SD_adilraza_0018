```python
name = input("Enter the  name: ")
course = input("Enter the course: ")
branch = input("Enter the branch: ")
rollnum = int(input("Enter the roll number: "))
marks_coa = int(input("Enter number of coa: "))
marks_ds = int(input("Enter number of ds: "))
marks_dstl = int(input("Enter number of dstl: "))
marks_m4 = int(input("Enter number of m4: "))
marks_uhv = int(input("Enter number of uhv: "))
marks_python = int(input("Enter number of python: "))
print("name:", name)
print("course:", course)
print("branch:", branch)
print("roll number: ", rollnum)
if marks_coa >= 90 and marks_coa <= 100:
  print("coa grade A")
elif marks_coa < 90:
  print("coa grade B")
if marks_ds >= 90 and marks_ds <= 100:
  print("ds grade A")
elif marks_ds < 90:
  print("ds grade B")
if marks_dstl >= 90 and marks_dstl <= 100:
  print("dstl grade A")
elif marks_dstl < 90:
  print("dstl grade B")
if marks_m4 >= 90 and marks_m4 <= 100:
  print("m4 grade A")
elif marks_m4 < 90:
  print("m4 grade B")
if marks_uhv >= 90 and marks_uhv <= 100:
  print("uhv grade A")
elif marks_uhv < 90:
  print("uhv grade B")
if marks_python >= 90 and marks_python <= 100:
  print("python grade A")
elif marks_python < 90:
  print("python grade B")

sum1 = marks_ds + marks_coa + marks_dstl + marks_m4 + marks_python + marks_uhv
per = (sum1 / 600) * 100
print("Percentage = ",per)
```

```python
l = [1,2,3,4,4,"sudh","kmumar",324,34.456,"abc"]
```

```python
l1_num = []
l2_str = []
for i in l:
  if type(i) == int or type(i) == float:
    l1_num.append(i)
  else:
    l2_str.append(i)
```

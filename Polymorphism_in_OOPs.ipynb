```python
class data_science:
  def syllabus(self):
    print("this is my syllabus for data science masters ")
```

```python
class web_dev:
  def syllabus(self):
    print("this is my syllabus for web dev")
```

```python
def class_parcer(class_obj):
  for i in class_obj:
    i.syllabus()
```

```python
data_science = data_science()
```

```python
web_dev = web_dev()
```

```python
class_obj = [data_science , web_dev]
```

```python
class_parcer(class_obj)
```
